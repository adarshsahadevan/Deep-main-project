#### OpenCV is a huge open-source library for computer vision , machine learning ,and image processing .OpenCV It supports programming languages like Python, C++, and Java,etc.
#### Image & Video Processing: OpenCV can process images and videos to identify objects, faces, or even handwriting.
#### Integration with NumPy: When combined with libraries like NumPy, a library optimized for numerical operations, the capabilities of OpenCV are significantly enhanced.
#### Computer Vision: It is essential for understanding and manipulating images and videos, serving as a foundation for Artificial Intelligence applications such as self-driving cars, robotics, and photo correction apps.

#### There are 3 types of images:
#### Gray scaled, coloured, black & white.
#### coloured has 3 layers , red ,green and blue (each ranges to 0 to 255 pixel).
#### gray scaled has one layer.
#### black & white has 2 layers , black and white.

```python
image = cv2.imread('dog.jpg')

cv2.imshow('dog_imagefit',image)
cv2.waitKey(0)       # is used to display an image in a window for a specified amount of time.
cv2.destroyAllWindows() # to close all the windows that were created by the cv2.imshow() function.
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
# image resize

image = cv2.imread('dog.jpg')

img = cv2.resize(image,(550,700))

import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
img.shape # here 3 means 3 layered (red,green,blue).
```

```python
# Grayscale conversion
img = cv2.imread('dog.jpg',cv2.IMREAD_GRAYSCALE)

import matplotlib.pyplot as plt 
plt.imshow(img)
```

```python
img.shape # in this it is one layered so there is 1.
```
