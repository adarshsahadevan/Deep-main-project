2. Develop a text summarizer using a Transformer model to concisely capture key
points from news articles while ensuring factual accuracy and readability.

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn", tokenizer="facebook/bart-large-cnn")

def summarize_article(article_text, max_length=100, min_length=50):
    article_text = article_text.strip()
    summary = summarizer(article_text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']
```

```python
# sample article
article = "Space exploration has come a long way since the first moon landing in 1969. NASA Artemis mission is set to return humans to the moon by 2025, with a focus on sustainable lunar exploration. The mission will involve new technologies, such as advanced space suits and lunar habitats. Additionally, international cooperation is playing a significant role in this endeavor. Countries like Canada and Japan are contributing to the mission by providing essential components like the lunar Gateway and the lunar rover. Experts believe this mission will pave the way for the first human expedition to Mars in the coming decades."
```

```python
summary = summarize_article(article)

print("\nOriginal Article:\n", article)
print("\nSummary:\n", summary)
```
