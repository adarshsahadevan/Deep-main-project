Question: You are tasked with building a simple face detection system using OpenCV's
pre-trained Haarcascade Frontal Face classifier. The system should be able to detect
faces in a given image. Complete the following steps:
1. Load an image that contains one or more human faces (you can use a sample
image from the web or your own collection).
2. Use OpenCV to read the image and convert it to grayscale.
3. Load the haarcascade_frontalface_default.xml file and apply the Haarcascade
classifier to detect faces in the image.
4. Draw bounding boxes around the detected faces and display the final image with
the bounding boxes.
5. Save the output image with the detected faces highlighted

```python
import cv2
```

```python
img_color = cv2.imread('/content/drive/MyDrive/Softroniics/Tests/Test Deep Learning/face3.webp')
```

```python
img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(img_gray, scaleFactor=1.1, minNeighbors=5)
```

```python
for (x, y, w, h) in faces:
        cv2.rectangle(img_color, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
# Import the cv2_imshow function from google.colab.patches
from google.colab.patches import cv2_imshow

# Replace cv2.imshow with cv2_imshow
cv2_imshow(img_color)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('image.jpg', img_color)
```
